```python
print("Welcome to HealthIntel CLI!")
print("Track your daily health data easily.\n")

# Collect user information
name = input("Enter your name: ")
age = int(input("Enter your age: "))

print(f"\nHello {name}, age {age}. Let's begin tracking your health data.\n")

# Collect basic health details
print("Please enter your basic health details below.\n")

height = float(input("Enter your height (in cm): "))
weight = float(input("Enter your weight (in kg): "))
goal = input("Enter your goal (Fitness / Weight Loss / Weight Gain): ")

# Display the collected summary
print("\n--- Health Summary ---")
print(f"Name: {name}")
print(f"Age: {age}")
print(f"Height: {height} cm")
print(f"Weight: {weight} kg")
print(f"Goal: {goal}")
print("\nYour information has been recorded successfully.")
```
