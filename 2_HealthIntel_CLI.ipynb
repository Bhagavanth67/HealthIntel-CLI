```python
print("Welcome back to HealthIntel CLI!")
print("Today, weâ€™ll analyze your BMI (Body Mass Index).\n")

# Step 1: Collect basic details
name = input("Enter your name: ")
age = int(input("Enter your age: "))
height = float(input("Enter your height (in m): "))
weight = float(input("Enter your weight (in kg): "))

# Step 2: Calculate BMI
bmi = weight / (height ** 2)

# Step 3: Determine BMI Category
if bmi < 18.5:
    category = "Underweight"
elif 18.5 <= bmi < 24.9:
    category = "Healthy weight"
elif 25 <= bmi < 29.9:
    category = "Overweight"
else:
    category = "Obese"

# Step 4: Display Results
print("\n--- BMI Analysis Report ---")
print(f"Name: {name}")
print(f"Age: {age}")
print(f"Height: {height} cm")
print(f"Weight: {weight} kg")
print(f"BMI: {bmi:.2f}")
print(f"Category: {category}")

# Step 5: Health Tips based on BMI
print("\n--- Health Recommendation ---")

if category == "Underweight":
    print("You are under the recommended weight range. Try including more nutritious calories and proteins.")
elif category == "Healthy weight":
    print("Great job! You are in a healthy range. Maintain your diet and regular activity.")
elif category == "Overweight":
    print("You are slightly above the healthy range. Consider light exercise and balanced meals.")
else:
    print("Your BMI is in the obese range. Regular physical activity and dietary management are advised.")

print("\nAnalysis complete. Stay consistent with your health goals.")
```
